```python
import pandas as pd
```

```python
df = pd.read_csv('../sourceData/PSUold.csv')
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
tmpPower = []
tmpPrice = []
value = []
for i in range(len(df['power'])):
    try:
        df['power'][i] = df['power'][i].replace('Вт', '')
        df['power'][i] = df['power'][i].replace('?', '0')
        df['price'][i] = df['price'][i].replace('?', '0')
        df['power'][i] = int(df['power'][i])
    except Exception as e:
        print(f'Line {i} ERROR: {e}')
    tmpPower.append(df['power'][i])
    tmpPrice.append(df['price'][i])
    if int(tmpPower[i]) != 0 and int(tmpPrice[i]) != 0:
        value.append(int(tmpPower[i]) / int(tmpPrice[i]))
    else:
        value.append('0')

df['value'] = value

df.head()
```

```python
df.to_csv('../data/PSU.csv')
```
