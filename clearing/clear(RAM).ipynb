```python
import pandas as pd
import numpy as np
```

```python
df_ram = pd.read_csv('../sourceData/RAMdry.csv')
df_ram
```

```python
df_ram.shape
```

```python
df_ram = df_ram[df_ram['price'] != '?']
df_ram.shape
```

```python
df_ram = df_ram[df_ram['timings'] != '?']
df_ram.shape
```

```python
df_ram = df_ram.drop(columns="Unnamed: 0")
df_ram.shape
```

```python
df_ram
```

```python
for i in range(len(df_ram['name'].values)):
    df_ram['freq'].values[i] = int(df_ram['freq'].values[i])
    df_ram['price'].values[i] = int(df_ram['price'].values[i])
tmp = []
for i in range(len(df_ram['name'].values)):
    tmp.append(int(df_ram['freq'].values[i]) / int(df_ram['price'].values[i]) + int(df_ram['capacity'].values[i]) / int(df_ram['price'].values[i]))
df_ram['value'] = tmp

df_ram.head()
```

```python
df_ram.to_csv('../data/RAM.csv')
```
