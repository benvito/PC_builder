```python
import pandas as pd
import numpy as np
```

```python
df_gpu = pd.read_csv('../sourceData/GPU_benchmarks_v7.csv')
df_gpu
```

```python
df_gpu = df_gpu[~np.isnan(df_gpu['price'])]
df_gpu.shape
```

```python
df_gpu
```

```python
df_gpu = df_gpu.drop(columns=['powerPerformance'])
```

```python
df_gpu
```

```python
for price in range(len(df_gpu['price'].values)):
    df_gpu['price'].values[price] = round(df_gpu['price'].values[price] * 80, 2)
```

```python
tmp = []
nvidia = ['GeForce', 'Quadro', 'ION', 'NVS', 'Tesla']
for i in range(len(df_gpu['gpuName'].values)):
    #if nvidia in df_gpu['gpuName'].values[i]:
    if any(j in df_gpu['gpuName'].values[i] for j in nvidia):
        tmp.append('NVIDIA')
    else:
        tmp.append('AMD')
df_gpu['brnd'] = tmp
df_gpu.head()
```

```python
df_gpu.to_csv('../data/GPU.csv')
```
