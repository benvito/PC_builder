```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('../data/GPU.csv')
df
```

```python
tmp = df[df['TDP'].isna()]
tmp
```

```python
df_train = df[~np.isnan(df['TDP'])]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df_train[['G3Dmark', 'G2Dmark', 'price']]
y = df_train['TDP']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X_new = [[5000, 2500], [6000, 3000]]
tdp_predict = model.predict(X)
print(tdp_predict)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, tdp_predict)
print("Коэффициент детерминации R2 = {:.2f}".format(r2))
```

```python
for tdp_i in range(len(df['TDP'].values)):
    try:
        df['TDP'].values[tdp_i] = round(float(df['TDP'].values[tdp_i]), 1)
        trued = True
    except:
        trued = False
    if trued:
        X_to_pred = [[df['G3Dmark'].values[tdp_i], df['G2Dmark'].values[tdp_i], df['price'].values[tdp_i]]]
        tdp_predicted = model.predict(X_to_pred)
        df['TDP'].values[tdp_i] = tdp_predicted
```
