```python

import pandas as pd
import numpy as np
```

```python
sockets = ['LGA 1200', 'LGA 1700', 'AM4', 'AM5', 'LGA 2066', 'LGA 1151v2']
```

```python
df_cpu = pd.read_csv('sourceData/CPU_benchmark_v4.csv')
df_cpu
```

```python
df_cpu = df_cpu[df_cpu['socket'] != 'unknown']
df_cpu.shape
```

```python
df_cpu = df_cpu[~np.isnan(df_cpu['price'])]
df_cpu.shape
```

```python
df_cpu = df_cpu.drop(columns=['testDate', 'powerPerf'])
```

```python
df_cpu
```

```python
for price in range(len(df_cpu['price'].values)):
    df_cpu['price'].values[price] = round(df_cpu['price'].values[price] * 80, 2)
```

```python
import regex as re
```

```python
for socket in range(len(df_cpu['socket'].values)):
    if re.match(r"^FC", df_cpu['socket'].values[socket]):
        reg = re.match(r"^FC", df_cpu['socket'].values[socket])
        df_cpu['socket'].values[socket] = df_cpu['socket'].values[socket][reg.end():]
    reg2 = re.match(r"[L|B]GA", df_cpu['socket'].values[socket])
    if reg2:
        df_cpu['socket'].values[socket] = df_cpu['socket'].values[socket][0:reg.end()+1] + ' ' + df_cpu['socket'].values[socket][reg.end()+1:]
```
