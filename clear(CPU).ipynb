```python
import pandas as pd
import numpy as np
```

```python
df_cpu = pd.read_csv('sourceData/CPU_benchmark_v4.csv')
df_cpu
```

```python
df_cpu = df_cpu[df_cpu['socket'] != 'unknown']
df_cpu.shape
```

```python
df_cpu = df_cpu[~np.isnan(df_cpu['price'])]
df_cpu.shape
```

```python
df_cpu = df_cpu.drop(columns=['testDate', 'powerPerf'])
```

```python
df_cpu
```

```python
for price in range(len(df_cpu['price'].values)):
    df_cpu['price'].values[price] = round(df_cpu['price'].values[price] * 80, 2)
```

```python
df_cpu.to_csv('data/CPU.csv')
```
